```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np


array_2d = np.array([
    [0, 6, 4, 4],  # Jarak A ke A, B, C, D
    [6, 0, 8, 6],  # Jarak B ke A, B, C, D
    [4, 8, 0, 8],  # Jarak C ke A, B, C, D
    [4, 6, 8, 0]   # Jarak D ke A, B, C, D
])
print("Array 2D:")
print(array_2d)
z = linkage(array_2d)
plt.figure(figsize=(25,10))
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("sample index")
plt.ylabel("distance")
dendrogram(z, leaf_rotation=90)
plt.show()
```
