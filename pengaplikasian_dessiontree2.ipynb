```python
# Mengimpor pustaka yang diperlukan
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
from google.colab import files

# Mengunggah file CSV
uploaded = files.upload()

# Membaca file CSV
df = pd.read_csv('/content/drive/MyDrive/archive/Iklan_sosmed.csv')

# Menampilkan 5 baris pertama dari DataFrame
print(df.head())

# Mengubah variabel kategorikal menjadi numerik
df['Kelamin'] = df['Kelamin'].map({'Laki': 0, 'Perempuan': 1})

# Memisahkan fitur dan target
X = df[['Kelamin', 'Usia', 'EstimasiGaji']]  # Fitur
y = df['Beli']  # Target

# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Membuat model Decision Tree
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Memprediksi
y_pred = model.predict(X_test)

# Menampilkan hasil prediksi
print("Prediksi:", y_pred)

# Menghitung akurasi
accuracy = model.score(X_test, y_test)
print("Akurasi:", accuracy)

# Visualisasi pohon keputusan
plt.figure(figsize=(12,8))
tree.plot_tree(model, feature_names=X.columns, class_names=['Tidak Beli', 'Beli'], filled=True)
plt.show()
```
