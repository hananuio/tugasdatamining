```python
import numpy as np
import matplotlib.pyplot as plt

# Data titik siswa (x, y)
data = {
    "M1": (1, 4.5),
    "M2": (3, 6.5),
    "M3": (4, 4.5),
    "M4": (7.5, 3.2),
    "M5": (6, 2.3),
    "M6": (2.5, 3.8),
    "M7": (5, 5.5),
}

# Titik pusat awal cluster
C1 = (3, 4)
C2 = (6, 4)

# Fungsi untuk menghitung jarak Euclidean
def euclidean_distance(point1, point2):
    return np.sqrt((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2)

# Mengelompokkan data ke dalam cluster berdasarkan jarak terdekat
cluster_C1 = []
cluster_C2 = []

for key, value in data.items():
    distance_to_C1 = euclidean_distance(value, C1)
    distance_to_C2 = euclidean_distance(value, C2)

    if distance_to_C1 < distance_to_C2:
        cluster_C1.append((key, value))
    else:
        cluster_C2.append((key, value))

# Menampilkan hasil klasterisasi
print("Cluster 1 (C1):", [point[0] for point in cluster_C1])
print("Cluster 2 (C2):", [point[0] for point in cluster_C2])

# Visualisasi
plt.figure(figsize=(8, 6))

# Plot data klaster C1
for point in cluster_C1:
    plt.scatter(point[1][0], point[1][1], color='blue', label='Cluster 1 (C1)' if 'Cluster 1 (C1)' not in plt.gca().get_legend_handles_labels()[1] else "")

# Plot data klaster C2
for point in cluster_C2:
    plt.scatter(point[1][0], point[1][1], color='red', label='Cluster 2 (C2)' if 'Cluster 2 (C2)' not in plt.gca().get_legend_handles_labels()[1] else "")

# Plot titik pusat cluster
plt.scatter(C1[0], C1[1], color='cyan', label='C1 (3,4)', marker='x', s=100)
plt.scatter(C2[0], C2[1], color='magenta', label='C2 (6,4)', marker='x', s=100)

# Memberikan label pada setiap titik data
for key, value in data.items():
    plt.text(value[0] + 0.1, value[1], key, fontsize=9)

# Pengaturan grafik
plt.title('Visualisasi Klasterisasi')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
